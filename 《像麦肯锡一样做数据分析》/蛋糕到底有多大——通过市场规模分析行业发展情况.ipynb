# 蛋糕到底有多大——通过市场规模分析行业发展情况

### 一、 市场规模计算：现有规模&未来规模预测
#### 市场规模测算的两种方法
1. Top Down：自上而下
- 偏于宏观层面的测算
- 基于对整体发展的考量
- 测算速度相对较快，单精度相对较低
2. Bottom Up：自下而上
- 考虑该市场的各个细分市场
- 该市场的各类驱动因素
- 测算更加精确，过程相对复杂  
实际应用中，会两种方法结合使用，使测算准确度更高  
得出来的数据不一定是完全一样的，至少要在一个数量级上，这样能确保市场预测的准确度较高
判断市场的规模和未来的发展速度  

注意：  
- 通过对市场进行测算，了解市场的整体规模和预测未来的增速，为企业制定战略和发展目标奠定基础和方向
- 没有完全准确的市场测算，都是基于事实的推测

#### 自上而下
- 大市场到小市场：从全球市场到亚太市场到中国市场
- 与宏观发展要素相关：利用GDP/人口/进出口总额等宏观数据，进行回归分析预测
- 同类对标：如美国欧洲日本邓世昌类似发展路径时期的规模  

优势：
- 测算方式相对简单
- 测算结果得出较快  

劣势：
- 需要相关数据支持，需要考虑可得性
- 方法较为粗糙，缺少细节支撑  

适用场景：
- 已知比所求市场更大范围市场规模（如全球市场）或宏观数据
- 需要快速了解一个市场的量级

方法示例公式1：
1. 大市场到小市场的计算方法：已知大市场 * 目标市场占比
2. 宏观数据或发展趋势判断：
    >1.适用于人口、经济等宏观要素强关联性的市场预测  
     2.国际其他市场相对比较成熟，可以提供借鉴

方法示例公式2：
    目标市场 = 已知规模 + 预期规模
    预期规模 = 某年某次发生时间的次数 * 预计未来年复合增长率
    未来年复合增长率考虑因素：
    > 整个行业历史增长率 + 行业中专业人员对行业未来的预测
    > 目前行业中单个企业或机构为代表，看其单体规模的增长
    > 若供给端较充足，则看需求端市场价位的接受程度或渗透率
    

自上而下方法案例1：
   > 如何估算中国未来（2020年）旅游人次的数量2020年旅游人次 = 2020年人口总量 * 人均年旅游出行次数
   
   计算2020年人口总量所需数据：
   > 中国人口现状
   > 中国人口未来变化趋势及2020年预测数据
   
   数据来源：
   > 中国统计年鉴
   > 中国十三五规划
   
   计算人均年旅游出行次数：
   > 当前我国每年的旅行人次
   > 美国人均GDP与旅游人数对应关系
   > 中国人均GDP及预测

自上而下方法案例2：
    > 测算北京机场旅客数量：当前旅客吞吐量 + 新增数量
    > 新增数量 = 北京当地人数 * 航空出行频次 + 周边人口 * 航空出行频次
    
   方法及数据来源：
   >民航局统计公报（包含历年详细吞吐量数据）<br/>
    北京市统计局、统计公报<br/>
    考虑背景辐射范围较广，将周边人口定义为北京周边接壤城市，除去天津（有自己的机场）
   
   
#### 自下而上
- 细分市场加总:将市场拆解成A.B.C.D等若干小市场或区域，在注意进行测算
- 市场强相关因素：如高铁长度、用电规模、旅游人数
- 发展驱动因素：考虑渗透率、价格、生产要素等核心因素  

优势：
- 测算细致，准确度相对较高
- 可以对局部结果进行微调  

劣势：
- 耗时较长
- 需要较多假设作为前提条件  

适用场景：
- 需要细致测算行业发展趋势和背后原因的测算（如各个细分市场的分别趋势）
- 当缺少宏观或整体数据信息的时候：

自下而上方法示例公式：  
  &emsp; 目标市场= 目标用户群数量 * 目标渗透率 * 客单价<br>
  &emsp; 目标用户群数量 = 目标用户群数量 * 目标渗透率 * 客单价<br>
  &emsp; 目标用户群数量的确定：  
  &emsp;&emsp; 用户细分：瞄准的是哪一类特殊用户？  
  &emsp;&emsp; 产品细分：真正的产品可能是什么？  
  <br>
   &emsp; 影响目标渗透率的因素：  
   - 刚需程度
   - 发展阶段
   - 竞争方案（产品的竞争力如何）
   - 外部环境
    <br>
   &emsp; 影响客单价的因素：要搞清楚定价的方式是什么
   - 模块：只是产品价格还是包括有其他服务的价格
   - 成本加成：原材料、利润等
   - 企业间的博弈：通过降低价格、成本获取客户
   
**自下而上方法案例：估算中国线上母婴市场规模**  
线上母婴市场规模 = 母婴用户总数量 * 线上渗透率 * 年下单数量 * 客单价 
- 母婴用户总数量：中国2017年新生儿数量
- 线上渗透率：
- 年下单数量：
- 平均每单花费：

**测算当前市场和未来市场的异同**
相同点：都可以用自上而下和自下而上两种方法测算
不同点：
- 关注用户数量是否变化
- 关注渗透率是否变化
- 关注下单数是否变化
- 关注客单价是否变化

### 二、 四步搭建市场规模预测框架
四步法：
|四步法|将回答的问题|
|------|-----------|
|1. 列出行业分析框架|如何搭建有效的行业分析框架|
|2. 按照分析框架列出历史数据|历史数据从哪里获得|
|3. 对未来3-5年相关变量进行预测|未来的预测数据从哪里获得<br>
对未来规模进行预测时应该注意哪些要点|
|4. 对市场预测模型进行归纳、思考|如何进行最有效、直观的结果展示|

#### 第一步：列出行业分析框架
方法1：客户渗透率分解法
>定义：由客群基数出发，将所需预测的指标拆解为客群人数 * 相关渗透率

举例1：预测2019-2024中国居民天然气的市场规模
影响中国居民天然气市场的因素：
- 中国居民家庭户数：总家庭数 * 城镇化率
- 天然气渗透率：
- 用气价格：平均用气数量 * 平均用气价格

客户渗透率分解法适用情景：
&emsp; 产品本身有可替代性（不是领域内唯一），也就是说在其市场中不处于垄断地位，渗透率是重要且可得的

方法2：细分行业品类法
举例2预测2019-2024年中国汽车销售额
中国汽车行业销售额 = 燃油车销售额 + 油电混合车销售额 + 电动汽车销售额

细分行业品类适用情景：
适用范围较广，无论是行业本身具有细分领域，或者行业本身可以分成几类不用的 应用场景，都可使用

*框架搭建要点*
1. 金字塔原则  
    框架层次要清晰，符合常识
    拆分要符合MECE原则  
    MECE原则：拆分出来的市场彼此独立，不会出现包含和被包含的关系，但是各个拆分出来的市场组合起来又是一个完整的市场
    
2. 80/20原则  
       尽量抓住核心主要市场，适当舍弃

3. 灵活调整
- 在无法了解准确数字的时候，可以采用对标（如看相关同行业年报）、专家访谈、门店寻访等方式进行推测
- 如果在后期发现数据不可得，需要及时调整框架

#### 第二步：按照分析框架，列出历史数据

*一般数据来源:*

|数据分类|数据来源|
|----|----|
|原始可用数据|国家颁布的统计数据:<br>国家统计局（人口普查）<br>工信部（产业）<br>xxx白/蓝皮书（政策）<br>发展计划书等|
|一手数据|市场走访<br>抽样调查<br>问卷整理<br>焦点小组<br>暗访<br>cold call等|
|二手数据|市场调研公司<br>专家访谈所得数据整理<br>投行和券商研究部数据等|

*行业数据来源：*

|行业分类|该行业常见数据来源|
|----|----|
|医药行业|IMS咨询|

#### 第三步：进行未来3-5年相关变量的预测
1. 相关技巧和应注意的要点  
&emsp; 在分析未来规模时，一般会分为不同场景，如：消极，基础，乐观  
&emsp; 针对未来预测，有Top down和Bottom up两种方法，通常资讯公司在搭建模型时，会用两种方法分别计算，交叉验证

2. 未来的数据从哪里获得
回归分析法：

|方法|案例|思路|
|----|----|----|
|回归分析法|预测2018-2023年中国贷款余额|思路：一般GDP与贷款余额线性相关<br>1.整体2008-2018年GDP的贷款余额数据<br>2.求出2008-2018年GDP与贷款余额的线性相关关系<br>3.预测2018-2023年GDP数据（可采用IMF等机构的专业预测）<br>4.求解2018-2023年的贷款余额数据|
|对标法|预测2018-2023年中国汽车保有总量|通过“人口数量 * 人均汽车保有量”的思路对该指标进行预测<br><br>其中，对于“人均汽车保有量”这个指标，对标美国，欧盟，日本，韩国等发达经济体，进行预估，如“2023年达到发达经济体人均指标的70%”|
|专家访谈或相关报告|电动车销量在2023年会占到汽车总销量的比重|1. 对交通部政策司的相关专家进行访谈，了解其内部的规划和对未来形势的判断<br>2. 参考券商，咨询公司专门的电动车行业研究报告，寻找相关数据，并做交叉对比|

#### 第四步：对市场预测模型进行归纳、总结
1. 列出方法论和分析框架
2. 描述核心结果和理论
3. 列出核心假设
